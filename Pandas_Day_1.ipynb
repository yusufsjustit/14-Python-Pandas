# loc and iloc

- df.loc[rows, columns name(s)]
- df.iloc[rows, column index(es)]

```python
df.loc[0:34 ,"name"]
# return all rows but only name column
```

```python
df.loc[0:4 ,"name"]
# return first 5 rows but only name column
```

```python
df.loc[0:4, ["name","class"]]
```

```python
df.loc[:, "name"]
# return all rows
# dont write start and end of the range :
```

```python
df.loc[:10, "name"]
# return up to index 10
# dont write start and end of the range :
```

```python
df.loc[10:, ["name", "id"]]
# return from index 10
# dont write start and end of the range :
```

```python
df.head()
```

- id is index 0
- name is index 1
- class is index 2

```python
df.iloc[:, 1]
#1 is index for name column
```

```python
df.iloc[:, [1,2]]
#1 is index for name column
#2 is index for class
```

```python
# df.loc[0:4, ["name","class"]] first 5 rows
# in loc, stopping range is included in the output

df.iloc[0:5, [1,2]]
# in iloc, stopping range is excluded in the output
```

## how to filter

- using slicing method df[condition]
- boolean filtering

```python
df["mark"]>90
# return true or false
```

```python
df[df["mark"]>90]
```

```python
df[df["class"]=="Six"]
```

- filter by multiple conditions

```python
df[(df["class"]=="Six") & (df["gender"]=="male")]
```

```python
df[(df["gender"]=="female") & (df["mark"]>=85)]
```

```python
df[(df["gender"]=="female") | (df["mark"]>=90)]
# | this is called pipe
```

- df.loc[condition, column labels]

```python
df.loc[df["id"]<5, : ]
```

```python
df.loc[df["gender"]=="male", ["name","class"]]
```

```python
df.iloc[df["gender"]=="male", 1]
# iloc cant be used to filter
```
